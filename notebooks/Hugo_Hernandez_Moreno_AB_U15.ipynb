## **Ingesta de CSV's**


> Función para combinar varios archivos CSV ubicados en una carpeta en un solo DataFrame y guardar el resultado en un nuevo archivo CSV.





```python
import pandas as pd
import os
import numpy as np

def combinar_csv(ruta_carpeta, nombre_archivo_salida):
    lista_archivos = [f for f in os.listdir(ruta_carpeta) if f.endswith('.csv')]

    df_embalses = pd.DataFrame()

    for archivo in lista_archivos:
        ruta_archivo = os.path.join(ruta_carpeta, archivo)
        nombre_embalse = os.path.splitext(archivo)[0]

        try:
            df_temporal = pd.read_csv(ruta_archivo, sep=";", encoding='latin-1')
        except UnicodeDecodeError:
            try:
                df_temporal = pd.read_csv(ruta_archivo, sep=";", encoding='ISO-8859-1')
            except Exception as e:
                print(f"Could not read file {ruta_archivo} with common encodings: {e}")
                continue


        df_temporal['embalse'] = nombre_embalse

        df_embalses = pd.concat([df_embalses, df_temporal], ignore_index=True)


    df_embalses.to_csv(nombre_archivo_salida, index=False)

ruta_carpeta = '/content/drive/MyDrive/Classroom/1ºPHE(MARZO) CYBERSECURITY APPLIED COMPUTING UNIT 15 - ARTIFICIAL INTELLIGENCE & INTELLIGENT SYSTEMS/AB/csv'
nombre_archivo_salida = "mixembalses.csv"

combinar_csv(ruta_carpeta, nombre_archivo_salida)
```

## **Prophet**

```python
import plotly.graph_objects as go
from prophet import Prophet
import pandas as pd

meses_dict = {
    'enero': '01', 'febrero': '02', 'marzo': '03', 'abril': '04',
    'mayo': '05', 'junio': '06', 'julio': '07', 'agosto': '08',
    'septiembre': '09', 'octubre': '10', 'noviembre': '11', 'diciembre': '12'
}

# Diccionario para almacenar modelos y predicciones
resultados_prediccion = {}

for embalse, df_embalse in dfs_por_embalse.items():
    df_prophet = df_embalse.copy()

    # Convertir mes en número
    df_prophet['month_num'] = df_prophet['month'].map(meses_dict).astype(str)

    # Crear la columna ds (formato fecha)
    df_prophet['ds'] = pd.to_datetime(df_prophet['year'].astype(str) + '-' + df_prophet['month_num'] + '-01')

    df_prophet['y'] = df_prophet['cub_hec']
    df_prophet = df_prophet[['ds', 'y']].dropna()

    # Entrenar el modelo
    model = Prophet(yearly_seasonality=True, daily_seasonality=False, weekly_seasonality=False)
    model.fit(df_prophet)

    # Fechas futuras (12 años)
    future = model.make_future_dataframe(periods=12*12, freq='MS')
    forecast = model.predict(future)

    # Guardar resultados
    resultados_prediccion[embalse] = {
        'model': model,
        'forecast': forecast,
        'historical': df_prophet
    }

    print(f"Predicción completada para {embalse}")
```

```python
# Graficamos la serie temporal de Prophet
for embalse, resultado in resultados_prediccion.items():
    historical = resultado['historical']
    forecast = resultado['forecast']

    fig = go.Figure()

    # Serie histórica
    fig.add_trace(go.Scatter(
        x=historical['ds'],
        y=historical['y'],
        mode='lines+markers',
        name='Histórico',
        line=dict(color='blue')
    ))

    # Predicción
    fig.add_trace(go.Scatter(
        x=forecast['ds'],
        y=forecast['yhat'],
        mode='lines',
        name='Predicción',
        line=dict(color='green', dash='dash')
    ))

    # Intervalo de confianza
    fig.add_trace(go.Scatter(
        x=forecast['ds'],
        y=forecast['yhat_lower'],
        mode='lines',
        name='Límite inferior',
        line=dict(color='lightgreen'),
        showlegend=False
    ))

    fig.add_trace(go.Scatter(
        x=forecast['ds'],
        y=forecast['yhat_upper'],
        mode='lines',
        name='Límite superior',
        fill='tonexty',
        line=dict(color='lightgreen'),
        showlegend=False
    ))

    # Layout
    fig.update_layout(
        title=f'Predicción de volumen embalsado - {embalse}',
        xaxis_title='Fecha',
        yaxis_title='Hectómetros cúbicos',
        template='plotly_white',
        height=500,
        legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1)
    )

    fig.show()
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

y_true_total = []
y_pred_total = []

for embalse, resultado in resultados_prediccion.items():
    df_real = resultado['historical']
    df_pred = resultado['forecast']

    # Comparar solo en el histórico
    df_merge = pd.merge(df_real, df_pred[['ds', 'yhat']], on='ds', how='inner')

    y_true_total.extend(df_merge['y'].values)
    y_pred_total.extend(df_merge['yhat'].values)

# Convertir a arrays
y_true_total = np.array(y_true_total)
y_pred_total = np.array(y_pred_total)

# Métricas
r2_total = r2_score(y_true_total, y_pred_total)
mae = mean_absolute_error(y_true_total, y_pred_total)
rmse = np.sqrt(mean_squared_error(y_true_total, y_pred_total))

# MAPE corregido
non_zero_mask = y_true_total != 0
mape = np.mean(np.abs((y_true_total[non_zero_mask] - y_pred_total[non_zero_mask]) / y_true_total[non_zero_mask])) * 100

# Mostramos métricas
print(f"\nMétricas globales del modelo Prophet:")
print(f"R²: {r2_total:.4f}")
print(f"MAE: {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"MAPE: {mape:.2f}%")
print(f"Accuracy aproximada: {100 - mape:.2f}%")
```

```python
from prophet import Prophet
import pandas as pd

# Preparar datos
meses_dict = {
    'enero': '01', 'febrero': '02', 'marzo': '03', 'abril': '04',
    'mayo': '05', 'junio': '06', 'julio': '07', 'agosto': '08',
    'septiembre': '09', 'octubre': '10', 'noviembre': '11', 'diciembre': '12'
}

df_fechas = df_pivot.copy()
df_fechas['month_num'] = df_fechas['month'].map(meses_dict).astype(str)
df_fechas['fecha'] = pd.to_datetime(df_fechas['year'].astype(str) + '-' + df_fechas['month_num'] + '-01')

columnas_embalses = [col for col in df_fechas.columns if col not in ['year', 'month', 'month_num', 'fecha']]
df_fechas['total'] = df_fechas[columnas_embalses].sum(axis=1, skipna=True)

# Renombrar para Prophet
fechas = df_fechas[['fecha', 'total']].rename(columns={'fecha': 'ds', 'total': 'y'})
fechas = fechas.dropna()

# Entrenar modelo
model = Prophet()
model.fit(fechas)

# Generar predicciones 12 años
proximas_fechas = model.make_future_dataframe(periods=144, freq='MS')
predicciones = model.predict(proximas_fechas)

# Separar predicciones futuras
predicciones_futuras = predicciones[predicciones['ds'] > fechas['ds'].max()]

print("Modelo Prophet entrenado")
print(f"Datos historicos: {len(fechas)} registros")
print(f"Predicciones futuras: {len(predicciones_futuras)} registros")
```

## **SARIMAX**

Predicción de 12 años por embalse, no total de agua, sino dividida en embalses.

```python
import plotly.graph_objects as go
from statsmodels.tsa.arima.model import ARIMA
import pandas as pd

resultados_arima = {}

for embalse, df in dfs_por_embalse.items():
    df_model = df.copy()

    meses_dict = {
        'enero': '01', 'febrero': '02', 'marzo': '03', 'abril': '04',
        'mayo': '05', 'junio': '06', 'julio': '07', 'agosto': '08',
        'septiembre': '09', 'octubre': '10', 'noviembre': '11', 'diciembre': '12'
    }
    df_model['month_num'] = df_model['month'].map(meses_dict).astype(str)
    df_model['ds'] = pd.to_datetime(df_model['year'].astype(str) + '-' + df_model['month_num'] + '-01')
    df_model = df_model[['ds', 'cub_hec']].set_index('ds').sort_index()

    df_model = df_model.asfreq('MS')

    # Entrenar ARIMA simple (p,d,q)
    model = ARIMA(df_model, order=(1,1,1), seasonal_order=(1,1,1,12))
    model_fit = model.fit()

    # Predicción a 12 años (144 meses)
    forecast = model_fit.get_forecast(steps=144)
    pred_ci = forecast.conf_int()

    # Fechas futuras para el forecast
    future_dates = pd.date_range(start=df_model.index[-1] + pd.DateOffset(months=1), periods=144, freq='MS')

    # Guardamos resultados
    resultados_arima[embalse] = {
        'model': model_fit,
        'historical': df_model,
        'forecast': forecast.predicted_mean,
        'conf_int': pred_ci,
        'future_dates': future_dates
    }

    fig = go.Figure()

    # Datos históricos
    fig.add_trace(go.Scatter(
        x=df_model.index, y=df_model['cub_hec'],
        mode='lines', name='Histórico'
    ))

    # Predicción
    fig.add_trace(go.Scatter(
        x=future_dates, y=forecast.predicted_mean,
        mode='lines', name='Predicción ARIMA'
    ))

    # Intervalo de confianza
    fig.add_trace(go.Scatter(
        x=list(future_dates) + list(future_dates[::-1]),
        y=list(pred_ci.iloc[:,0]) + list(pred_ci.iloc[:,1][::-1]),
        fill='toself',
        fillcolor='rgba(0,100,80,0.2)',
        line=dict(color='rgba(255,255,255,0)'),
        hoverinfo="skip",
        showlegend=True,
        name='Intervalo confianza'
    ))

    fig.update_layout(
        title=f"Predicción SARIMA para {embalse}",
        xaxis_title='Fecha',
        yaxis_title='Hectómetros cúbicos',
        template='plotly_white',
        height=500
    )

    fig.show()
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

y_true_total = []
y_pred_total = []

for embalse, resultados in resultados_arima.items():
    df_hist = resultados['historical']  # Serie histórica con índice datetime
    pred = resultados['model'].predict(start=df_hist.index[0], end=df_hist.index[-1])

    # Guardamos reales y predichos para el histórico
    y_true_total.extend(df_hist['cub_hec'].values)
    y_pred_total.extend(pred.values)

# Convertimos a numpy arrays para métricas
y_true_total = np.array(y_true_total)
y_pred_total = np.array(y_pred_total)

# Calculamos métricas
r2_total = r2_score(y_true_total, y_pred_total)
mae_total = mean_absolute_error(y_true_total, y_pred_total)
rmse_total = np.sqrt(mean_squared_error(y_true_total, y_pred_total))

# Para MAPE evitamos división por cero
non_zero_idx = y_true_total != 0
mape_total = np.mean(np.abs((y_true_total[non_zero_idx] - y_pred_total[non_zero_idx]) / y_true_total[non_zero_idx])) * 100

accuracy = 100 - mape_total

print(f"R² total global SARIMA: {r2_total:.4f}")
print(f"MAE total SARIMA: {mae_total:.4f}")
print(f"RMSE total SARIMA: {rmse_total:.4f}")
print(f"MAPE total SARIMA: {mape_total:.4f}%")
print(f"Accuracy aproximada SARIMA: {accuracy:.2f}%")
```

## **LSTM**

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Bidirectional
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras import Input
import numpy as np
import pandas as pd
import random
import plotly.graph_objects as go
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

print("LSTM TOTAL")
print()

# Fijar seeds para reproducibilidad
tf.random.set_seed(42)
np.random.seed(42)
random.seed(42)

# Funciones auxiliares (iguales a tu código)
def create_dataset(features, targets, window_size=24):
    X, y = [], []
    for i in range(len(targets) - window_size):
        X.append(features[i:i+window_size])
        y.append(targets[i+window_size])
    return np.array(X), np.array(y)

def predict_model(model, X):
    return model(X, training=False)

# Preparar datos totales del sistema
meses_dict = {
    'enero': 1, 'febrero': 2, 'marzo': 3, 'abril': 4,
    'mayo': 5, 'junio': 6, 'julio': 7, 'agosto': 8,
    'septiembre': 9, 'octubre': 10, 'noviembre': 11, 'diciembre': 12
}

# Crear dataset total desde df_pivot
df_lstm_total = df_pivot.copy()
df_lstm_total['month_num'] = df_lstm_total['month'].map(meses_dict).astype(int)
df_lstm_total['ds'] = pd.to_datetime(df_lstm_total['year'].astype(str) + '-' +
                                    df_lstm_total['month_num'].astype(str) + '-01')

# Sumar todos los embalses
columnas_embalses = [col for col in df_lstm_total.columns if col not in ['year', 'month', 'month_num', 'ds']]
df_lstm_total['total_sistema'] = df_lstm_total[columnas_embalses].sum(axis=1, skipna=True)

# Preparar para LSTM
df_lstm_total = df_lstm_total[['ds', 'total_sistema', 'month_num']].set_index('ds').sort_index()


# Features cíclicas de mes
df_lstm_total['month_sin'] = np.sin(2 * np.pi * df_lstm_total['month_num'] / 12)
df_lstm_total['month_cos'] = np.cos(2 * np.pi * df_lstm_total['month_num'] / 12)

# Escalado del target
scaler_total = MinMaxScaler()
df_lstm_total['total_scaled'] = scaler_total.fit_transform(df_lstm_total[['total_sistema']])

# Crear lag features
for lag in range(1, 4):
    df_lstm_total[f'lag_{lag}'] = df_lstm_total['total_scaled'].shift(lag)

# Eliminar NaN
df_lstm_total = df_lstm_total.dropna()

# Escalar features
scaler_feats_total = MinMaxScaler()
df_lstm_total[['month_sin', 'month_cos', 'lag_1', 'lag_2', 'lag_3']] = scaler_feats_total.fit_transform(
    df_lstm_total[['month_sin', 'month_cos', 'lag_1', 'lag_2', 'lag_3']])

print(f"Features preparadas: {len(df_lstm_total)} registros válidos")

# Preparar matrices
features_total = df_lstm_total[['total_scaled', 'month_sin', 'month_cos', 'lag_1', 'lag_2', 'lag_3']].values
targets_total = df_lstm_total['total_scaled'].values

# Configuración de ventana
window_size = 24

# Crear datasets
X_total, y_total = create_dataset(features_total, targets_total, window_size)

# División train/val/test 70/15/15
train_end = int(len(X_total) * 0.7)
val_end = int(len(X_total) * 0.85)

X_train_total = X_total[:train_end]
y_train_total = y_total[:train_end]
X_val_total = X_total[train_end:val_end]
y_val_total = y_total[train_end:val_end]
X_test_total = X_total[val_end:]
y_test_total = y_total[val_end:]

print()
print(f"División de datos:")
print(f"- Entrenamiento: {len(X_train_total)} muestras")
print(f"- Validación: {len(X_val_total)} muestras")
print(f"- Test: {len(X_test_total)} muestras")

# Modelo Bidirectional LSTM

model_total = Sequential([
    Input(shape=(window_size, features_total.shape[1])),
    Bidirectional(LSTM(64, return_sequences=True)),
    Dropout(0.3),
    Bidirectional(LSTM(32)),
    Dropout(0.3),
    Dense(1)
])

model_total.compile(optimizer='adam', loss='mse')

early_stop = EarlyStopping(monitor='val_loss', patience=25, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=10, verbose=1)


# Entrenar modelo
history = model_total.fit(
    X_train_total, y_train_total,
    epochs=150,
    batch_size=16,
    validation_data=(X_val_total, y_val_total),
    callbacks=[early_stop, reduce_lr],
    verbose=0
)

y_pred_scaled_total = predict_model(model_total, X_test_total).numpy()
y_pred_total_final = scaler_total.inverse_transform(y_pred_scaled_total)
y_test_orig_total = scaler_total.inverse_transform(y_test_total.reshape(-1, 1))

# Fechas para test
test_dates_total = df_lstm_total.index[window_size + val_end: window_size + val_end + len(y_test_total)]

# Métricas
r2_lstm_total = r2_score(y_test_orig_total, y_pred_total_final)
mae_lstm_total = mean_absolute_error(y_test_orig_total, y_pred_total_final)
rmse_lstm_total = np.sqrt(mean_squared_error(y_test_orig_total, y_pred_total_final))
mape_lstm_total = np.mean(np.abs((y_test_orig_total - y_pred_total_final) / y_test_orig_total)) * 100
accuracy_lstm_total = 100 - mape_lstm_total

print(f"Métricas del modelo LSTM TOTAL:")
print(f"R²: {r2_lstm_total:.4f}")
print(f"MAE: {mae_lstm_total:.2f}")
print(f"RMSE: {rmse_lstm_total:.2f}")
print(f"MAPE: {mape_lstm_total:.2f}%")
print(f"Accuracy: {accuracy_lstm_total:.2f}%")

# Comparación con otros modelos
if 'mape_total' in locals() and 'mape_sarimax' in locals():
    print(f"\nComparación de modelos:")
    print(f"Prophet MAPE: {mape_total:.2f}%")
    print(f"SARIMAX MAPE: {mape_sarimax:.2f}%")
    print(f"LSTM MAPE: {mape_lstm_total:.2f}%")

    mejor_modelo = min([
        ('Prophet', mape_total),
        ('SARIMAX', mape_sarimax),
        ('LSTM', mape_lstm_total)
    ], key=lambda x: x[1])

    print(f"Mejor modelo: {mejor_modelo[0]} (MAPE: {mejor_modelo[1]:.2f}%)")

# Guardar resultados
resultados_lstm_total = {
    'model': model_total,
    'scaler': scaler_total,
    'scaler_feats': scaler_feats_total,
    'X_test': X_test_total,
    'y_test': y_test_total,
    'y_pred': y_pred_total_final,
    'y_test_orig': y_test_orig_total,
    'test_dates': test_dates_total,
    'r2': r2_lstm_total,
    'mae': mae_lstm_total,
    'rmse': rmse_lstm_total,
    'mape': mape_lstm_total,
    'accuracy': accuracy_lstm_total,
    'df_lstm': df_lstm_total,
    'history': history
}

print()
fig = go.Figure()

# Datos reales
fig.add_trace(go.Scatter(
    x=test_dates_total,
    y=y_test_orig_total.flatten(),
    mode='lines+markers',
    name='Real',
    line=dict(color='blue', width=2)
))

# Predicción LSTM
fig.add_trace(go.Scatter(
    x=test_dates_total,
    y=y_pred_total_final.flatten(),
    mode='lines+markers',
    name='Predicción LSTM',
    line=dict(color='red', width=2, dash='dash')
))

fig.update_layout(
    title=f"LSTM Total del Sistema - Predicción vs Real<br>R²={r2_lstm_total:.4f} - Accuracy={accuracy_lstm_total:.2f}%",
    xaxis_title='Fecha',
    yaxis_title='Capacidad Total (hm³)',
    template='plotly_white',
    height=600
)

fig.show()
```

## **Predicción 12 Años**

```python
n_meses = 144
window_size = 24

# Obtener datos del modelo entrenado
model = resultados_lstm_total['model']
scaler = resultados_lstm_total['scaler']
scaler_feats = resultados_lstm_total['scaler_feats']
df_lstm = resultados_lstm_total['df_lstm']
X_test = resultados_lstm_total['X_test']
test_dates = resultados_lstm_total['test_dates']

# Última fecha de los datos
last_date = test_dates[-1]

# Inicializar predicción
future_preds_scaled = []
current_window = X_test[-1].copy()
current_date = last_date

# Bucle de predicción iterativa
for i in range(n_meses):
    # Predecir siguiente mes
    pred_scaled = model.predict(current_window[np.newaxis, :, :], verbose=0)[0, 0]
    future_preds_scaled.append(pred_scaled)

    current_date += pd.DateOffset(months=1)
    next_month_num = current_date.month

    next_month_sin = np.sin(2 * np.pi * next_month_num / 12)
    next_month_cos = np.cos(2 * np.pi * next_month_num / 12)

    # Preparar lags actualizados
    lag_1 = pred_scaled
    lag_2 = current_window[-1, 0]
    lag_3 = current_window[-2, 0]

    # Crear DataFrame para escalar features
    feats_df = pd.DataFrame([[next_month_sin, next_month_cos, lag_1, lag_2, lag_3]],
                            columns=['month_sin', 'month_cos', 'lag_1', 'lag_2', 'lag_3'])

    # Escalar features
    feats_scaled = scaler_feats.transform(feats_df)[0]

    # Crear nueva fila para la ventana
    new_row = np.array([pred_scaled] + list(feats_scaled))

    # Actualizar ventana deslizante
    current_window = np.vstack([current_window[1:], new_row])

# Desescalar predicciones
future_preds = scaler.inverse_transform(np.array(future_preds_scaled).reshape(-1, 1)).flatten()

# Crear fechas futuras
future_dates = pd.date_range(
    start=last_date + pd.DateOffset(months=1),
    periods=n_meses,
    freq='MS'
)

try:
    historico_start = future_dates[0] - pd.DateOffset(years=5)
    historico = df_lstm.loc[df_lstm.index >= historico_start].copy()
    if len(historico) > 0:
        historico['total_original'] = scaler.inverse_transform(historico[['total_scaled']])
    else:
        historico = df_lstm.tail(60).copy()
        historico['total_original'] = scaler.inverse_transform(historico[['total_scaled']])
except:
    historico = df_lstm.tail(60).copy()
    historico['total_original'] = scaler.inverse_transform(historico[['total_scaled']])

fig = go.Figure()

# Datos históricos (últimos 5 años)
fig.add_trace(go.Scatter(
    x=historico.index,
    y=historico['total_original'].values,
    mode='lines+markers',
    name='Histórico (5 años)',
    line=dict(color='blue', width=2),
    marker=dict(size=4)
))

# Predicciones futuras
fig.add_trace(go.Scatter(
    x=future_dates,
    y=future_preds,
    mode='lines+markers',
    name='Predicción 12 años',
    line=dict(color='red', width=3, dash='dash'),
    marker=dict(size=4)
))


fig.add_annotation(
    x=last_date,
    y=historico['total_original'].iloc[-1] if len(historico) > 0 else future_preds[0],
    text="Inicio<br>Predicción",
    showarrow=True,
    arrowhead=2,
    arrowcolor="gray",
    bgcolor="white",
    bordercolor="gray"
)

fig.update_layout(
    title=f"Predicción LSTM: Sistema Total de Embalses - 12 años<br>"
          f"Desde {future_dates[0].strftime('%B %Y')} hasta {future_dates[-1].strftime('%B %Y')}",
    xaxis_title="Fecha",
    yaxis_title="Capacidad Total (hm³)",
    template="plotly_white",
    height=600,
    hovermode='x unified'
)

# Mostrar gráfico
fig.show()

prediccion_12_años = {
    'fechas': future_dates,
    'predicciones': future_preds,
    'predicciones_escaladas': future_preds_scaled,
    'modelo': model,
    'scaler': scaler,
    'historico': historico,
    'metricas_modelo': {
        'r2': resultados_lstm_total['r2'],
        'mape': resultados_lstm_total['mape'],
        'accuracy': resultados_lstm_total['accuracy']
    }
}
```
